## Тэйлор

```python
import math
ITTERATIONS=20;
def coss(x):
    x_pow=x
    result=1
    fact=1
    for i in range(1, ITTERATIONS):
        x_pow=(-1)**i*x**(2*i)
        fact*=2*i-1
        fact*=2*i
        result+=x_pow/fact
    return(result)
print(coss(0.4))
print(math.cos(0.4))
```

```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'svg')
import matplotlib.pyplot as plt
import numpy as np
vs = np.vectorize(coss)
print(coss, vs)

angles = np.r_[-16.25:16.25:0.01]
plt.plot(angles, np.cos(angles))
plt.plot(angles, vs(angles))
plt.show()
```
